## Data Preprocessing

```python
# Convert date to datetime format
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Create time feeatues
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Month_Name"] = df["Date"].dt.strftime("%b")
df["Weekday"] = df["Date"].dt.day_name()
```

```python
# Convert text columns to category
cols = ["Region", "District", "Channel", "Product"]

for col in cols:
    df[col] = df[col].astype("category")
```

## Create Business KPIs

```python
# Revenue per Unit
df["Revenue_per_Unit"] = df["Revenue_MWK"] / df["Units_Sold"]
```

```python
# Commission Rate
df["Commission_Rate"] = df["Commission_MWK"] / df["Revenue_MWK"]
```

```python
# Sales Value Category
df["Sales_Category"] = pd.cut(
    df["Revenue_MWK"],
    bins=[0, 10000, 50000, 100000, 500000],
    labels=["Low", "Medium", "High", "Very High"]
)
```

## AGGREGATE the Data (for Excel & Power BI)

```python
# Aggregation
df_summary = df.groupby(
    ["Year", "Month", "Region", "Channel", "Product"]
).agg({
    "Revenue_MWK": "sum",
    "Units_Sold": "sum",
    "Commission_MWK": "sum"
}).reset_index()
```

```python
# Add KPI to summary
df_summary["Commission_Rate"] = (
    df_summary["Commission_MWK"] / df_summary["Revenue_MWK"]
)
```

## Save the Two Datasets

```python
# Full clean dataset
df.to_csv("sales_clean.csv", index=False)

# Small dataset (for Excel & Power BI)
df_summary.to_csv("sales_summary.csv", index=False)
```
